## Football Premier League Project - 01 

```python
import pandas as pd
import numpy as np
df =pd.read_csv("results.csv")

```

```python
df= df.loc[(df['season']=='2006-2007')] #limited data acc to 2006-2007
```

```python
df
```

```python
team = list(df.home_team.unique()) #converting unique value into a list , to add data in new column of new table
```

```python
table = {'Club': team } #creating lists of dictionary 
```

```python
final_results = pd.DataFrame(table) #creating new dataframe
```

```python
Total_MP = list(df.home_team.value_counts() + df.away_team.value_counts())
final_results['MP'] = Total_MP
```

```python
wins = (pd.concat([df.loc[df['result']=='H']['home_team'] , df.loc[df['result']=='A']['away_team']]).value_counts()) #win counts
```

```python
draw = (pd.concat([df.loc[df['result']=='D']['home_team'] , df.loc[df['result']=='D']['away_team']]).value_counts()) #draw counts
```

```python
loss = (pd.concat([df.loc[df['result']=='A']['home_team'] , df.loc[df['result']=='H']['away_team']]).value_counts()) #Lose counts
```

```python
Goals_h = df.groupby(by=["home_team"])["home_goals"].sum()
Goals_a = df.groupby(by=["away_team"])["away_goals"].sum()
```

```python
Goals_for = (Goals_a + Goals_h)
```

```python
Goals_Against = (df.groupby(by=["home_team"])["away_goals"].sum() + df.groupby(by=["away_team"])["home_goals"].sum())
```

```python
Goals_difference = (Goals_for - Goals_Against)
```

```python
points = (3*wins + draw)
```

Adding Tables (win , loss , draw etc) into new column of new table with joints , 'On left' = 'Club' (Column name) we have club and inserting from right . So right_index = True. Merge Function will join two tables based on some column data. 

```python
final_results = final_results.merge(wins.rename('W') , left_on='Club', right_index=True )
final_results = final_results.merge(draw.rename('D') , left_on='Club', right_index=True )
final_results = final_results.merge(loss.rename('L') , left_on='Club', right_index=True )
final_results = final_results.merge(Goals_for.rename('GF') , left_on='Club', right_index=True )
final_results = final_results.merge(Goals_Against.rename('GA') , left_on='Club', right_index=True )
final_results = final_results.merge(Goals_difference.rename('GD') , left_on='Club', right_index=True )
final_results = final_results.merge(points.rename('Pts') , left_on='Club', right_index=True )
```

```python
final_results = final_results.sort_values('Pts' , ascending=False)
```

```python
final_results.reset_index(inplace = True, drop = True)
final_results
```
